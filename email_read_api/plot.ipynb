```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("data.csv")

# Remove entries with Chinese characters in the date column
df = df[~df['Date'].str.contains('週|今日')]

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y %H:%M')

# Extract the hour component from the datetime column
df['Hour'] = df['Date'].dt.hour

# Separate the data for MAFW and SECOM
mawf_data = df[df['Subject'] == 'MAWF']
secom_data = df[df['Subject'] == 'SECOM']

# Create separate bar charts for MAFW and SECOM events
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Plot MAFW events
ax[0].hist(mawf_data['Hour'], bins=24, color='blue', label='MAWF')
ax[0].set_title('MAWF Error')
ax[0].set_xlabel('Hour')
ax[0].set_ylabel('Count')
ax[0].set_xticks(np.arange(0, 25, 3))
ax[0].set_yticks(np.arange(0, max(ax[0].get_ylim()) + 1, 5))

# Plot SECOM events
ax[1].hist(secom_data['Hour'], bins=24, color='orange', label='SECOM')
ax[1].set_title('SECOM Errro')
ax[1].set_xlabel('Hour')
ax[1].set_ylabel('Count')
ax[1].set_xticks(np.arange(0, 25, 3))
ax[1].set_yticks(np.arange(0, max(ax[1].get_ylim()) + 1, 5))

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("data.csv")

# Remove entries with Chinese characters in the date column
df = df[~df['Date'].str.contains('週|今日')]

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y %H:%M')

# Extract the hour component from the datetime column
df['Hour'] = df['Date'].dt.hour

# Separate the data for MAFW and SECOM
mawf_data = df[df['Subject'] == 'MAWF']
secom_data = df[df['Subject'] == 'SECOM']

# Count occurrences of each hour for MAFW and SECOM
mawf_hour_counts = mawf_data.groupby('Hour').size()
secom_hour_counts = secom_data.groupby('Hour').size()

# Create scatter plot for both MAFW and SECOM events
plt.figure(figsize=(10, 6))

# Plot MAFW events
plt.scatter(mawf_hour_counts.index, mawf_hour_counts.values, s=100, color='blue', alpha=0.5, label='MAWF')

# Plot SECOM events
plt.scatter(secom_hour_counts.index, secom_hour_counts.values, s=100, color='orange', alpha=0.5, label='SECOM')

# Customize plot
plt.title('MAWF & SECOM Errors')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("data.csv")

# Remove entries with Chinese characters in the date column
df = df[~df['Date'].str.contains('週|今日')]

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y %H:%M')

# Extract the weekday component from the datetime column (Monday=0, Sunday=6)
df['Weekday'] = df['Date'].dt.weekday

# Map the weekday numbers to weekday names
weekday_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Weekday'] = df['Weekday'].map(lambda x: weekday_names[x])

# Specify the order of weekdays
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['Weekday'] = pd.Categorical(df['Weekday'], categories=weekday_order, ordered=True)

# Separate the data for MAFW and SECOM
mawf_data = df[df['Subject'] == 'MAWF']
secom_data = df[df['Subject'] == 'SECOM']

# Count occurrences of each weekday for MAFW and SECOM
mawf_weekday_counts = mawf_data.groupby('Weekday', observed=False).size()
secom_weekday_counts = secom_data.groupby('Weekday', observed=False).size()

# Create stacked bar chart for both MAFW and SECOM events
plt.figure(figsize=(10, 6))

# Plot MAFW events
plt.bar(mawf_weekday_counts.index, mawf_weekday_counts.values, color='blue', label='MAWF')

# Plot SECOM events on top of MAFW events
plt.bar(secom_weekday_counts.index, secom_weekday_counts.values, bottom=mawf_weekday_counts.values, color='orange', label='SECOM')

# Customize plot
plt.title('MAWF & SECOM Errors by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Count')
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv("data.csv")

# Remove entries with Chinese characters in the date column
df = df[~df['Date'].str.contains('週|今日')]

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y %H:%M')

# Extract the hour and weekday components from the datetime column
df['Hour'] = df['Date'].dt.hour
df['Weekday'] = df['Date'].dt.strftime('%A')  # Convert to full weekday name

# Separate the data for MAFW and SECOM errors
mawf_data = df[df['Subject'] == 'MAWF']
secom_data = df[df['Subject'] == 'SECOM']

# Create pivot tables to count occurrences of errors by hour and weekday for each error type
mawf_heatmap_data = mawf_data.pivot_table(index='Weekday', columns='Hour', aggfunc='size', fill_value=0)
secom_heatmap_data = secom_data.pivot_table(index='Weekday', columns='Hour', aggfunc='size', fill_value=0)

# Set the order of weekdays and hours
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
hour_order = range(24)
mawf_heatmap_data = mawf_heatmap_data.reindex(weekday_order, axis=0)
mawf_heatmap_data = mawf_heatmap_data.reindex(hour_order, axis=1)
secom_heatmap_data = secom_heatmap_data.reindex(weekday_order, axis=0)
secom_heatmap_data = secom_heatmap_data.reindex(hour_order, axis=1)

# Create the heatmaps
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# MAWF Errors Heatmap
sns.heatmap(mawf_heatmap_data, ax=axes[0], cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
axes[0].set_title('MAWF Errors')
axes[0].set_xlabel('Hour of the Day')
axes[0].set_ylabel('Weekday')

# SECOM Errors Heatmap
sns.heatmap(secom_heatmap_data, ax=axes[1], cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
axes[1].set_title('SECOM Errors')
axes[1].set_xlabel('Hour of the Day')
axes[1].set_ylabel('Weekday')

# Show plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv("data.csv")

# Remove entries with Chinese characters in the date column
df = df[~df['Date'].str.contains('週|今日')]

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y %H:%M')

# Extract the date component from the datetime column
df['Date'] = df['Date'].dt.date

# Group the data by date and subject, then count the occurrences
daily_counts = df.groupby(['Date', 'Subject']).size().unstack(fill_value=0)

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))

# Plot stacked bars for MAFW and SECOM events
plt.bar(daily_counts.index, daily_counts['MAWF'], color='blue', label='MAWF')
plt.bar(daily_counts.index, daily_counts['SECOM'], bottom=daily_counts['MAWF'], color='orange', label='SECOM')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Daily Count')
plt.title('Daily Count of Events')

# Add legend
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
